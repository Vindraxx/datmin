```python
# import library
import numpy as np
import pandas as pd
import statsmodels
import patsy
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# take a look at the dataset
df = pd.read_excel("Data Penjualan.xlsx")
df.head()
```

```python
# definition of research variables
X = df[['NilaiPenjualan']]
y = df['BiayaProduksi']
```

```python
X=sm.add_constant(X) #add a constant variable (Beta 0)
model=sm.OLS(y,X).fit()
model.summary()
```
